# Практическая работа после задания №6

```
class Building:
  def __init__(self,w,c,n=0):
    self.what=w
    self.color=c
    self.numbers=n
    self.mwhere(n)

  def mwhere(self,n):
    if n<=0:
      self.where="отстутствуют"
    elif 0<n<100:
      self.where = "малый склад"
    else:
      self.where="основной склад"

  def plus(self,p):
    self.numbers=self.numbers+p
    self.mwhere(self.numbers)
  def minus(self,m):
    self.numbers=self.numbers-m
    self.mwhere(self.numbers)

m1=Building("доски", "белые", 50)
m2=Building("доски", "коричневые", 300)
m3=Building("кирпичи","белые")

print (m1.what, m1.color, m1.where)
print (m2.what, m2.color, m2.where)
print (m3.what, m3.color, m3.where)

m1.plus(500)
print (m1.numbers, m1.where)
```

```
class pencil:
  def __init__ (self,c,l,n=0):
    self.color = c
    self.long = l
    self.numbers = n
    self.mwhere(n)

  def mwhere(self,n):
    if n<100:
      self.where = "детские"
    else:
      self.where = "школьные"
  
  def plus(self,p):
    self.numbers = self.numbers + p
    self.mwhere(self.numbers)
  def minus(self,m):
    self.numbers = self.numbers - m
    self.mwhere(self.numbers)

m1 = pencil("красные", "10", 25)
m2 = pencil("синие", "15", 156) 
m3 = pencil("желтые", "7")

print(m1.color, m1.long, m1.where)
print(m2.color, m2.long, m2.where)
print(m3.color, m3.long, m3.where)

m1.plus(100)
print(m1.numbers, m1.where)
```

# Практическая работа после задания №8

```
class Figure:
  color = "white"
  def changecolor(self, newcolor):
    self.color = newcolor

obj1 = Figure()

print (obj1.color)

obj1.changecolor("green")

print (obj1.color)

class oval(Figure):
  def __init__(self,l,h):
    self.long = l
    self.hight = h
  def outing(self):
    print(self.long, self.hight)

o1 = oval(2,5)
o1.outing()

class square(Figure):
  def __init__(self,l,h,w):
    self.long = l
    self.hight = h
    self.wight = w
  def outing(self):
    print(self.long, self.hight, self.wight)
s1 = square(2,4,10)
s1.outing()
```
