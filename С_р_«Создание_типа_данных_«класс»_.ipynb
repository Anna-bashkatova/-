```
class Worker: # класс Работник
  
  def __init__(self, year, days, name, oklad): # конструктор класса
    self.year=year
    self.days=365
    self.name=name
    
  def calc(self):  # функция подсчета стаж работы работника на данном предприятии 
    return 2020 - self.year
  
  def calc1(self): # функция по подсчету количества дней, которое прошло после года поступления на работу 
    return (2020 - self.year) * self.days

  def __del__(self): # деструктор 
    pass
  def __str__(self): # оператор преобразования в строку информации об бъекте
    return ('Фамилия: {}, оклад: {}, год поступления на работу: {}, \ncтаж работы работника на данном предприятии: {}, \nколичества дней, которое прошло после года поступления на работу: {}'.format(self.name, self.oklad, self.year, self.calc(), self.calc1()))
  
obj1 = Worker("","", "", "") # 1-ый обьект класса
obj1.name = input("Введите фамилию = ")
obj1.year = int(input("Введите год поступления на работу = "))
obj1.oklad = int(input("Введите оклад = "))

obj2 = Worker("","", "", "") # 2-ой обьект класса
obj2.name = input("Введите фамилию = ")
obj2.year = int(input("Введите год поступления на работу = "))
obj2.oklad = int(input("Введите оклад = "))

obj3 = Worker("","", "", "") # 3-ий обьект класса
obj3.name = input("Введите фамилию = ")
obj3.year = int(input("Введите год поступления на работу = "))
obj3.oklad = int(input("Введите оклад = "))

print(obj1)
print(obj2)
print(obj3)

print("Удаление объекта obj1")
print("Удаление объекта obj2")
print("Удаление объекта obj3")

del obj1
del obj2
del obj3
```
