```
class Persona: #Persona(персона) - родительский класс
  def __init__(self, n, s, a):
    self.name = n
    self.surname = s
    self.age = a

class Employee(Persona): #Класс наследника(служащий наследник персоны)
  def __init__(self, n, s, a, o):
    super().__init__(n,s,a) #Вызов родительского класса, применение наследования
    self.otdel = o

class Working(Employee):
  def __init__(self, n, s, a, o, p):
    super().__init__(n, s, a, o)
    self.produkt = p
  def zp(self, stavka):
    self.zp = self.produkt * int(stavka)

class Engineer(Employee):
  def __init__(self, n, s, a, o, t):
    super().__init__(n, s, a, o)
    self.time=t
  def zp(self, stavk):
    self.zp =  self.time * int(stavk)


w1 = Persona("Иван","Петров","32")
w2 = Employee("Сергей","Иванов","20","Бухгалтерия")
w3 = Working("Михаил","Сидоров","28","Цех-упаковывания", int(input("Введите количество изготовленных деалей рабочего:")))
w4 = Engineer("Алексей","Мешков","43","Инженерия", int(input("Введите количество отработанных часов инженера:")))
w3.zp(int(input("Введите ставку для подсчета зарплаты рабочего:")))
w4.zp(int(input("Введите ставку для подсчета зарплаты инженера:")))
```

```
print("Имя:", w1.name, "Фамилия:", w1.surname, "Возраст:", w1.age)
print("Имя:", w2.name, "Фамилия:", w2.surname, "Возраст:", w2.age, "Отдел:", w2.otdel)
print("Имя:", w3.name, "Фамилия:", w3.surname, "Возраст:", w3.age, "Отдел:", w3.otdel, "Количество произведенных продуктов:",w3.produkt, "Зарплата рабочего:", w3.zp)
print("Имя:", w4.name, "Фамилия:", w4.surname, "Возраст:", w4.age, "Отдел:", w4.otdel, "Количество отработанных часов:",w4.time,"Зарплата инженера:", w4.zp)
```
